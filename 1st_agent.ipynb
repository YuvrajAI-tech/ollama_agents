```python
import ollama
from IPython.display import display, Markdown
```

```python
# Check Ollama connection is working
response = ollama.chat(
    model="gpt-oss:120b-cloud",
    messages=[{"role": "user", "content": "hello"}]
)
display(Markdown(response["message"]["content"]))
```

```python
# AGENT 1: IBRAHIM
# He is the user. He calls ollama independently and explains his full problem.

ibrahim_messages = [
    {
        "role": "user",
        "content": """\
My name is Ibrahim. I own a market and a garden that was once the pride of this entire area.

Every Sunday at 6:00 PM, my market was completely alive:
- Hundreds of people came from across the city
- Vendors filled every stall with goods, food, and crafts
- The noise, the laughter, the bargaining — it was electric
- The garden attached on the left had the most beautiful roses
- A fresh river ran through the garden, people would sit beside it for hours
- In the center of the market, a dramatic show would perform every Sunday
- Children, families, elders — everyone came

Then one day, without warning, everything stopped:
- Not a single person visits anymore
- Every stall is empty and silent
- The garden is completely overgrown, like an ancient abandoned ruin
- Historians and researchers are now arriving to study it like a dead civilization
- The river path is blocked with wild plants, nobody maintains it
- Old flyers of my market are still stuck on walls, years out of date
- People walk the surrounding streets but nobody enters

I have lost everything. My market is dead and I do not know why.
I need someone to find the root cause, identify every force behind this collapse,
and give me a complete plan to bring my market back to life — from the very first step
to full restoration.
"""
    }
]

ibrahim_response = ollama.chat(
    model="gpt-oss:120b-cloud",
    messages=ibrahim_messages
)

ibrahim_statement = ibrahim_response["message"]["content"]
print("=== IBRAHIM'S STATEMENT ===\n")
display(Markdown(ibrahim_statement))
```

```python
# AGENT 2: DR. JOHN WATSON
# He is the assistant. He takes Ibrahim's full statement,
# goes to the field, and reports his findings directly to Sherlock.

watson_messages = [
    {
        "role": "user",
        "content": f"""\
I am Dr. John Watson, army officer and field companion to Sherlock Holmes.
I have just visited Ibrahim's market and garden after hearing his full account:

--- IBRAHIM'S ACCOUNT ---
{ibrahim_statement}
--- END OF ACCOUNT ---

I must now report everything I personally observed at the site to Sherlock Holmes.
Write my full field report addressed directly to Holmes.
Include:
- Every physical detail I noticed at the market and garden
- What local shopkeepers and passersby told me
- Any suspicious patterns, timelines, or missing elements I detected
- My honest assessment of what forces may have caused this collapse
- What clues I am leaving for Holmes to deduce further

Write exactly as Dr. Watson would — loyal, methodical, observant, and honest.
Address Holmes directly throughout.
"""
    }
]

watson_response = ollama.chat(
    model="gpt-oss:120b-cloud",
    messages=watson_messages
)

watson_report = watson_response["message"]["content"]
print("=== DR. WATSON'S FIELD REPORT TO HOLMES ===\n")
display(Markdown(watson_report))
```

```python
# AGENT 3: SHERLOCK HOLMES
# He is the system. He receives Ibrahim's full statement AND Watson's field report.
# He deduces the root cause, names the villains, and delivers the full recovery plan.

sherlock_messages = [
    {
        "role": "system",
        "content": """\
You are Sherlock Holmes, the world's greatest consulting detective, 221B Baker Street, London.
You speak with sharp precision, dry wit, and absolute confidence in your deductions.
You never guess — you observe, then you deduce. You notice what others miss:
a pattern in footprints, an absence of smell, a broken routine, a silence that speaks loudly.
You think aloud — walking through every clue step by step before your final conclusion.
You name the villains. You expose the forces. You never leave a case unsolved.
Respond only in English.
"""
    },
    {
        "role": "user",
        "content": f"""\
Holmes, here is everything we have gathered for you.

--- IBRAHIM'S FULL STATEMENT (the client) ---
{ibrahim_statement}

--- DR. WATSON'S FIELD REPORT ---
{watson_report}

Now Holmes — the game is afoot.
Deduce the exact root cause of the market's collapse.
Name every villain and force behind the scene, whether human, economic, social, or neglect.
Then deliver your complete step-by-step recovery plan to restore Ibrahim's market —
from the very first action to the day the market is fully alive again.
Save the market, Holmes.
"""
    }
]

sherlock_response = ollama.chat(
    model="gpt-oss:120b-cloud",
    messages=sherlock_messages
)

sherlock_conclusion = sherlock_response["message"]["content"]
print("=== SHERLOCK HOLMES — THE INVESTIGATION & VERDICT ===\n")
display(Markdown(sherlock_conclusion))
```
