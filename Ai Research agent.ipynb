```python
import ollama
from IPython.display import display, Markdown
```

```python
# Check everything is going on and fine to work next

response = ollama.chat(
    model = "glm-4.7:cloud",
    messages = [{"role": "user", "content": "hi, who are you"}]
)

display(Markdown(response["message"]["content"]))
```

```python
# Ask llm to Generate a Problem
question = "Define a systematic problem in AI's Training Database and how they accomplish a task with only it's training"
question += "Define problems in your conversation with user wants something and you can't give him right choice"
question += "Your Task is Defining ONLY PROBLEM NOT THE SOLUTION OR ANSWER OR ANY EXPLAINATION i don't need these extra thing anymore"
question += "Think you are a Ai Developer and you developed a ai and you finded some huge problem that can cause the conversation"

messages = [{"role": "user", "content": question}]
```

```python
#ask the llm to carry out problem with question messages
response = ollama.chat(
    model = "cogito-2.1:671b-cloud",
    messages=messages
)
question=response["message"]["content"]
display(Markdown(question))
```

```python
systematic_map = f"You are an ai develpor and you encounter some problems with your developed ai in conversation {question}, make decisions and deeply think of all problems and design a systematic map that can explain problem deeply with it's root causes"
systematic_map += f"Design a larg Map with systematic rules and workflows to determine problem and root causes of {question}"
systematic_map += f"ONLY Map out the problem and root cause of ai {question} dont include solution or answer. i only need a Map overview and right explanation or our {question}"

messages = [{"role": "user", "content": systematic_map}]
```

```python
response = ollama.chat(
    model = "qwen3-vl:235b-cloud",
    messages = messages
)

systematic_map = response["message"]["content"]
display(Markdown(systematic_map))
```

```python
plan = f"""you are a experienced expert of computer programming, data science, AI(ML/DL), NLP. 
you are a genius scientist in the entire world, you own your 3000+ patents. 
one day a ai developer called you and he say "i developed a ai autonomouse agent that can handle all tasks and behaviour like humans but some certain problems occur that i encounter in {question},
i designed a systematic map under the {systematic_map} to see what exactly going wrong with ai conversation."
you say "ok mr. i am comming to solve your problem and plus we make a environment that ensure in future this problem is never come again."
now you go to ai developer and meet him plus you see every aspect of he encounter in {question} + {systematic_map}.
you successfully solver his problem and make a environment that you promise with him during call time.
with above statement perform exactly that i told you and give exact plan to solution."""

messages = [{"role": "system", "content": plan}]
```

```python
response = ollama.chat(
    model = "cogito-2.1:671b-cloud",
    messages=messages
)

plan = response["message"]["content"]
display(Markdown(plan))
```

```python
solution = f"You are a expert of computer science and Programming/AI(ML/DL) you work with a ai developr and a scientist that own his 3000+ patents these both users encounter some problems and design a map under the {question}, {systematic_map}, and define a plan {plan} to execute the solution. your role is you research in the {question}, {systematic_map}, {plan} and after all you generate the perfect and fit all sides in this situation a solution, solution that you made for perfect work-able ai autonomuse ai agent. generate entire solution with the respect of {question + systematic_map + plan} and finish the whole criteria right now."
messages = [{"role": "user", "content": solution}]
```

```python
response = ollama.chat(
    model = "gpt-oss:120b-cloud",
    messages=messages
)

solution=response["message"]["content"]
display(Markdown(solution))
```

```python
execute = f""" You are a Expert computer programmer and software developer + AI developer.
Your task is develope a computer software that runs in linux with remaining and focus on {question}, {systematic_map}, {plan}, {solution}.
write code to develop your software """
messages = [{"role": "user", "content": execute}]
```

```python
response = ollama.chat(
    model = "minimax-m2.5:cloud",
    messages=messages
)

execute = response["message"]["content"]
display(Markdown(execute))
```
