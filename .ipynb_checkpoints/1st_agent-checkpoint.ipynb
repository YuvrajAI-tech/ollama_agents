```python
import ollama
from IPython.display import display, Markdown
```

```python
# check everything is going on or not
response = ollama.chat(
    model="gpt-oss:120b-cloud",
    messages=[{"role": "user", "content": "hello"}]
)

display(Markdown(response["message"]["content"]))
```

```python
#CALL TO AGENT1: DETECTIVE
response = ollama.chat(
    model="gpt-oss:120b-cloud",
    messages=[{"role": "detective", "content": market}]
)
detective = response["message"]["content"]
display(Markdown(detective))
```
